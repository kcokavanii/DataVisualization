## Импортируем необходимые библиотеки

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

## Прочитаем csv файл

```python
df = pd.read_csv('../data/ab-test.csv')
df
```

## Получим датафрейм до новостного потока

```python
df_before = df[df['time'] == 'before']
df_before
```

## Получим датафрейм после новостного потока

```python
df_after = df[df['time'] == 'after']
df_after
```

## Построим график

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 8))

df_after.boxplot(column='diff',
                by='group',
                ax = ax[0],
                color=dict(whiskers='red',
                           medians='red', caps='red'),
                boxprops=dict(color='green', linewidth=3),
                medianprops=dict(linewidth=2),
                showmeans=True)
df_before.boxplot(column='diff', 
                by='group',
                ax = ax[1],
                color=dict(whiskers='red',
                           medians='red', caps='red'),
                boxprops=dict(color='green', linewidth=3),
                medianprops=dict(linewidth=2),
                showmeans=True)
ax[0].set_title('after')
ax[1].set_title('before')
```

## Какой был IQR (интерквартильный размах) группы контроля до новостного потока?
Посчитаем его, посмотрев на график

**Ответ:** -10 - (-160) = **150**
